```python
 data1.columns = ['MR1']
data2.columns = ['MR2']
data3.columns = ['MR3']

# Zusammenfügen der DataFrames horizontal
merged_data = pd.concat([data1.head(5), data2.head(5), data3.head(5)], axis=1)

display(merged_data)
```

```python
import pandas as pd
from tabulate import tabulate

# Mittelwert MR1, MR2, MR3 
mw1 = round(data1['MR1'].mean(), 4)
mw2 = round(data2['MR2'].mean(), 4)
mw3 = round(data3['MR3'].mean(), 4)

# empirische Varianz MR1, MR2, MR3
var1 = round(data1['MR1'].var(), 4)
var2 = round(data2['MR2'].var(), 4)
var3 = round(data3['MR3'].var(), 4)

# empirische Standardabweichung MR1, MR2, MR3
std1 = round(data1['MR1'].std(), 4)
std2 = round(data2['MR2'].std(), 4)
std3 = round(data3['MR3'].std(), 4)

# Median MR1, MR2, MR3
m1 = round(data1['MR1'].median(), 4)
m2 = round(data2['MR2'].median(), 4)
m3 = round(data3['MR3'].median(), 4)

# Spannweite MR1, MR2, MR3
sw1 = round(data1['MR1'].max() - data1['MR1'].min(), 4)
sw2 = round(data2['MR2'].max() - data2['MR2'].min(), 4)
sw3 = round(data3['MR3'].max() - data3['MR3'].min(), 4)

# Ergebnis anzeigen 
ergebnisse = pd.DataFrame({
    'Kennzahlen': ['Mittelwert', 'Varianz', 'Std-Abw', 'Median', 'Spannweite'],
    'MR1': [mw1, var1, std1, m1, sw1],
    'MR2': [mw2, var2, std2, m2, sw2],
    'MR3': [mw3, var3, std3, m3, sw3]
})

ergebnisse
```

```python
#Aufgabe c)

# Erstellen der Histogramme für MR1, MR2, MR3
plt.figure(figsize=(15, 5))

# Histogramm für MR1
plt.subplot(1, 3, 1)
counts, bins, _ = plt.hist(data1['MR1'], bins=10, color='y', edgecolor='black')
plt.title('Histogramm MR1')

# Zahlen auf den Säulen anzeigen
for count, bin in zip(counts, bins):
    plt.text(bin + 0.5, count, str(int(count)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
#Aufgabe d)
#Histogramm MR1: Die Normalverteilung liegt vor da die  Vorrausetzungen erfüllt sind. Es hat Hochpunkt um den Mittelwert, und sie ist symmetrisch.
#Histogramm MR2:Die Normalverteilung liegt nicht, sie ist zwar symetrisch aber sie sollte einen Hochpunkt um den Mittelwert haben.
#Histogramm MR3: Eine Nomalverteilung liegt nicht vor, da es weder symetrisch noch einen Hochpunkt hat und sie um den Mittelwert ist.
```

```python
#Aufgabe e)
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Daten für MR1
data_mr1 = data1['MR1']

# Histogramm mit passender Normalverteilungsdichte für MR1
plt.figure(figsize=(8, 4))

# Histogramm für MR1 mit 10 Bins und Normalisierung auf Wahrscheinlichkeitsdichte
n, bins, _ = plt.hist(data_mr1, bins=10, density=True, alpha=0.6, color='b', edgecolor='black', label='Häufigkeit')

# Passende Normalverteilungsdichte für MR1
x = np.linspace(data_mr1.min(), data_mr1.max(), 100)
pdf = stats.norm.pdf(x, mw1, std1)
plt.plot(x, pdf, 'r-', label='angepasste NV_dichte')

plt.title('Histogramm und Normalverteilungsdichte für MR1')
plt.xlabel('x')
plt.ylabel('Anzahl')
plt.legend()
plt.show()

# Ausgabe der berechneten Parameter für MR1
print(f'Mittelwert (µ) für MR1: {mw1:.4f}')
print(f'Standardabweichung (σ) für MR1: {std1:.4f}')
```

```python
#Aufgabe g) 
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Verteilungsfunktion für MR1
cdf = stats.norm.cdf(x, mw1, std1)
plt.plot(x, cdf, 'g--', label='angepasste NV_Verteilungsfunktion')

plt.title('Verteilungsfunktion für MR1')
plt.xlabel('x')
plt.ylabel('p')
plt.legend()
plt.grid(axis='y')  # Horizontale Gitterlinien
plt.show()
```
