```python
import pandas as pd
frauen = df.loc[df['Anrede']=='Frau', ['Alter', 'Altersgruppe']]
frauen = frauen.reset_index(drop=True)


maenner = df.loc[df['Anrede']=='Herr', ['Alter', 'Altersgruppe']]
maenner = maenner.reset_index()
frauen['Alter']
```

```python
Häufigkeit = pd.Series(df['Altersgruppe']).value_counts(sort=False).rename_axis('Gruppe').reset_index(name='abs. Häufigkeit')
```

```python
Häufigkeit_frauen = pd.Series(frauen['Altersgruppe']).value_counts(sort=False).rename_axis('Gruppe Frauen').reset_index(name='abs. Häufigkeit')
Häufigkeit_frauen
```

```python
Häufigkeit_maenner = pd.Series(maenner['Altersgruppe']).value_counts(sort=False).rename_axis('Gruppe Männer').reset_index(name='abs. Häufigkeit')
Häufigkeit_maenner
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
x = np.arange(5) 
yfrauen = [1,12,14,15,12]  
ymaenner = [1,11,18,16,4]
width = 0.40
  
plt.bar(x-0.2, yfrauen, width) 
plt.bar(x+0.2, ymaenner, width) 

plt.xlabel('Schulnoten')
plt.ylabel('Häufigkeit')
plt.title('Altersverteilung Frauen (Blau) und Männer (Orange) zum Stichtag 31.12.2005')
plt.xticks(x, ['(0,20]', '(20,30]', '(30,40]', '(40,50]', '(50,200]'])
plt.show()
```

```python
#Total sum Frauen
total_sum_frauen = Häufigkeit_frauen['abs. Häufigkeit'].sum()
total_sum_frauen
```

```python
#Relative Häufigkeit Frauen
Häufigkeit_frauen['rel. Häufigkeit in %']= (Häufigkeit_frauen['abs. Häufigkeit']/total_sum_frauen*100).round(2)
Häufigkeit_frauen
```

```python
#plt.bar(Häufigkeit['Gruppe'], Häufigkeit['abs. Häufigkeit'])

plt.pie(Häufigkeit_frauen['rel. Häufigkeit in %'],labels= Häufigkeit_frauen['Gruppe Frauen'],autopct='%1.1f%%')
plt.title("Altersverteilung Frauen")
plt.show()
```

```python
#Total sum Männer
total_sum_maenner = Häufigkeit_maenner['abs. Häufigkeit'].sum()
total_sum_maenner
```

```python
#Relative Häufigkeit Männer
Häufigkeit_maenner['rel. Häufigkeit in %']= (Häufigkeit_maenner['abs. Häufigkeit']/total_sum_maenner*100).round(2)
```

```python
#plt.bar(Häufigkeit['Gruppe'], Häufigkeit['abs. Häufigkeit'])

plt.pie(Häufigkeit_maenner['rel. Häufigkeit in %'],labels= Häufigkeit_maenner['Gruppe Männer'],autopct='%1.1f%%')
plt.title("Altersverteilung Männer")
plt.show()
```
